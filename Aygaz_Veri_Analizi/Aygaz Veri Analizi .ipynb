```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

### Sütunları


id: Konaklama biriminin benzersiz tanımlayıcısı (unique identifier).

name: Konaklama biriminin adı.

host_id: Ev sahibinin benzersiz tanımlayıcısı (unique identifier).

host_name: Ev sahibinin ismi.

neighbourhood_group: New York şehri içindeki beş büyük bölgeden biri.

neighbourhood: New York şehri içindeki mahallelerden biri.

latitude: Konaklama biriminin enlem değeri.

longitude: Konaklama biriminin boylam değeri.

room_type: Konaklama biriminin tipi

price: Konaklama biriminin gecelik ücreti.

minimum_nights: En az konaklama süresi .

number_of_reviews: Toplam yapılmış yorum sayısı.

last_review: En son yapılmış yorum tarihi.

reviews_per_month: Aylık ortalama yorum sayısı.

calculated_host_listings_count: Ev sahibinin toplam ilan sayısı.

availability_365: Yıl boyunca müsaitlik durumu, 365 gün.

```python
neighborhood_groups = df.groupby('neighbourhood_group')['price'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='neighbourhood_group', y='price', data=neighborhood_groups)
plt.title('Mahalle Gruplarına Göre Ortalama Fiyatlar')
plt.xlabel('Mahalle Grupları')
plt.ylabel('Ortalama Fiyat')
plt.show()
```

```python

pivot_df = df.pivot_table(index='neighbourhood', columns='room_type', values='price', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, cmap='viridis', annot=True, fmt='.0f', linewidths=.5)
plt.title('Mahallelerdeki Oda Tiplerinin Ortalama Fiyatları')
plt.xlabel('Oda Tipi')
plt.ylabel('Mahalle')
plt.show()
```

```python

top_reviewed = df.nlargest(10, 'number_of_reviews')[['name', 'number_of_reviews']]
plt.figure(figsize=(10, 6))
sns.barplot(x='number_of_reviews', y='name', data=top_reviewed, orient='h')
plt.title('En Çok Yorum Alan 10 Konaklama Birimi')
plt.xlabel('Yorum Sayısı')
plt.ylabel('Konaklama Birimi Adı')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['minimum_nights'], bins=30, kde=False)
plt.title('Minimum Konaklama Süresinin Dağılımı')
plt.xlabel('Minimum Konaklama Süresi (Gece)')
plt.ylabel('Sayı')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['minimum_nights'], df['price'], alpha=0.5)
plt.title('Minimum Gece Konaklama ile Fiyat Arasındaki İlişki')
plt.xlabel('Minimum Gece Konaklama')
plt.ylabel('Fiyat')
plt.show()
```

```python

top_host_names = df['host_name'].value_counts().nlargest(5)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_host_names.index, y=top_host_names.values)
plt.title('En Çok Evi Olan 5 Kişi')
plt.xlabel('Ev Sahibi İsimleri')
plt.ylabel('Ev Sayısı')
plt.xticks(rotation=45)
plt.show()
```

```python
host_total_earnings = df.groupby("host_name")["price"].sum().nlargest(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=host_total_earnings.index, y=host_total_earnings.values)
plt.title('En Çok Kazanç Sağlayan 5 Ev Sahibi')
plt.xlabel('Ev Sahibi İsimleri')
plt.ylabel('Toplam Kazanç (Fiyat)')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='room_type', data=df)
plt.title('Oda Tiplerine Göre Konaklama Birimleri')
plt.xlabel('Oda Tipi')
plt.ylabel('Sayı')
plt.show()
```

```python

room_type_total_earnings = df.groupby("room_type")["price"].sum()


plt.figure(figsize=(8, 6))
plt.pie(room_type_total_earnings, labels=room_type_total_earnings.index, autopct='%1.1f%%', startangle=140)
plt.title('Oda Tiplerine Göre Toplam Kazanç Dağılımı')
plt.axis('equal')
plt.show()
```

```python
top_expensive_listings = df.nlargest(11, 'price')[['neighbourhood', 'name', 'price']]
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood', y='price', data=top_expensive_listings, ci=None)
plt.title('En Pahalı 10 Mahalle')
plt.xlabel('Mahalle')
plt.ylabel('Fiyat')
plt.xticks(rotation=45)
plt.show()
```

```python

top_reviewed_neighbourhoods = df['neighbourhood'].value_counts().nlargest(6).index


df_top_reviews = df[df['neighbourhood'].isin(top_reviewed_neighbourhoods)]

total_reviews_by_neighbourhood = df_top_reviews['neighbourhood'].value_counts()


plt.figure(figsize=(12, 8))
sns.barplot(x=total_reviews_by_neighbourhood.index, y=total_reviews_by_neighbourhood.values, ci=None)
plt.title('En Çok Yorum Alan 6 Mahalle')
plt.xlabel('Mahalle')
plt.ylabel('Toplam Yorum Sayısı')
plt.xticks(rotation=45)
plt.show()
```

```python

# En çok yorum alan 6 mahalleyi belirleyin
top_reviewed_neighbourhoods = df['neighbourhood'].value_counts().nlargest(6).index

# En çok yorum alan mahalleleri içeren yeni bir DataFrame oluşturun
df_top_reviews = df[df['neighbourhood'].isin(top_reviewed_neighbourhoods)]

# Mahallelerdeki konaklama birimlerinin ortalama fiyatlarını hesaplayın
average_prices_by_neighbourhood = df_top_reviews.groupby('neighbourhood')['price'].mean().reset_index()

# Görselleştirme için bar grafiği oluşturun
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood', y='price', data=average_prices_by_neighbourhood, ci=None)
plt.title('En Çok Yorum Alan Mahallelerdeki Ortalama Fiyatlar')
plt.xlabel('Mahalle')
plt.ylabel('Ortalama Fiyat')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_df_avail = df.pivot_table(index='neighbourhood', columns='room_type', values='availability_365', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df_avail, cmap='coolwarm', annot=True, fmt='.0f', linewidths=.5)
plt.title('Mahallelerin Toplam Müsaitlik Günleri')
plt.xlabel('Oda Tipi')
plt.ylabel('Mahalle')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['calculated_host_listings_count'], bins=30, kde=False)
plt.title('Ev Sahibinin Toplam İlan Sayısı Dağılımı')
plt.xlabel('Ev sahibinin toplam ilan sayısı')
plt.ylabel('Sayı')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calculated_host_listings_count', y='price', data=df)
plt.title('Ev Sahiplerinin İlan Sayısına Göre Toplam Kazançlar')
plt.xlabel('Ev Sahiplerinin İlan Sayısı')
plt.ylabel('Toplam Kazanç (Fiyat)')
plt.show()
```

```python

room_type_distribution = df.groupby(['neighbourhood_group', 'room_type']).size().unstack()


plt.figure(figsize=(12, 8))
room_type_distribution.plot(kind='bar', stacked=True)
plt.title('Mahalle Gruplarına Göre Oda Tipi Dağılımı')
plt.xlabel('Mahalle Grupları')
plt.ylabel('Sayı')
plt.legend(title='Oda Tipi', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
selected_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
selected_df = df[selected_columns]


correlation_matrix = selected_df.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Değişkenler Arası Korelasyon Matrisi')
plt.show()
```
